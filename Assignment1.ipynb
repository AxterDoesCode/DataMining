```python
import pandas as pd
from sklearn.datasets import load_wine

data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = pd.Series(data.target)
```

```python
data.target_names
```

```python
df.groupby('target').size()
```

1a) Target is the categorical feature, it is the value that corresponds to one of three values, 0, 1 and 2. The frequency of 0 is 59, the frequency of 1 is 71 and the frequency of 2 is 48.

1b) For univariate summaries I chose to calculate the standard deviation and the mean of each numerical feature

For multivariate summaries, I created a scatter matrix of the numerical features and a covariance matrix.

```python
display(df.drop('target', axis = 1).std())
```

```python
display(df.drop('target', axis = 1).mean())
```

```python
pd.plotting.scatter_matrix(df.drop('target', axis = 1))
df.drop('target', axis = 1).cov()
```

1c)

```python
df.groupby('target').median()
```

1d) First I need to find the pair with the highest correlation: This is flavanoids and total_phenols (below code shows the correlation matrix and then sorts the values from highest to lowest) The highest being 0.8645 which is the flavanoids and total_phenols pair.

```python
df.drop('target', axis = 1).corr()
```

```python
sorted_corr = df.drop('target', axis = 1).corr().stack().sort_values(ascending = False)
sorted_corr
```

```python
df.plot(x = 'flavanoids', y = 'total_phenols', kind = 'scatter')
```
